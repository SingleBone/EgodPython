```python
# import some packge
import torch
import numpy as np
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import torch.utils.data as Data
from torch import nn
```

```python
torch.manual_seed(1)
```

```python
# Hyper Parameters
TIME_STEP = 10
INPUT_SIZE = 1
LR = 0.02
DOWNLOAD_MNIST = False
```

```python
# RNN
class RNN(nn.Module):
    
    def __init__(self):
        super(RNN,self).__init__()
        
        self.rnn = nn.RNN(
            input_size=1,
            hidden_size=64,
            num_layers=1,
            batch_first=True,
        )
        self.out = nn.Linear(64,1)
        
    def forward(self,x,h_state):
        # x (batch, time_step, input_size)
        # h_state (n_layers, batch, hidden_size)
        # r_out (batch, time_step, output_size)
        r_out, h_state = self.rnn(x, h_state)   # h_state 也要作为 RNN 的一个输入
        outs = self.out(r_out.view(-1,64))
        return outs,h_state
rnn = RNN().to('cuda')
print(rnn)
```

```python
optimizer = torch.optim.Adam(rnn.parameters(),lr=LR)
loss_func = nn.MSELoss()
```

```python
h_state = None
```

```python
for step in range(100):
    start,end = step*np.pi,(step+1)*np.pi
    steps = np.linspace(start,end,10,dtype=np.float32)
    x_np = np.sin(steps)
    y_np = np.cos(steps)
    
    x = torch.from_numpy(x_np[np.newaxis,:,np.newaxis]).to('cuda')
    y = torch.from_numpy(y_np).to('cuda')
    
    prediction,h_state = rnn(x,h_state)
    h_state = h_state.data
    
    loss = loss_func(prediction,y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    plt.plot(steps, y_np.flatten(), 'r-')
    plt.plot(steps, prediction.to('cpu').data.numpy().flatten(), 'b-')
    plt.draw(); 
    plt.pause(0.05)
```

```python
a = np.linspace(0,2*np.pi,100,dtype=np.float32)
x = torch.from_numpy(np.sin(a)[np.newaxis,:,np.newaxis]).to('cuda')
y = np.cos(a)
h_state = None
y_pre,h_state = rnn.to('cuda')(x,h_state)
plt.plot(range(100),y,label='target')
plt.plot(range(100),np.sin(a),label='input')
plt.plot(range(100),y_pre.to('cpu').data.numpy(),label='prediction')
plt.legend(loc='best')
```
