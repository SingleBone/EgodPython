pandas处理丢失数据（NaN)

```python
df = DF(np.arange(24).reshape((6,4)),index=pd.date_range('20190301',periods=6),columns=['A','B','C','D'])
df.iloc[1,1]=np.nan
df.iloc[3,3]=np.nan
df
```

直接去掉含有NaN的行和列

```python
df1=df
df1.dropna(
    axis = 0, # 0是按行操作，1是按列操作
    how = 'any', # 'any'是只要有NaN就去掉，‘all'是全部为NaN才去掉
)
df1.dropna(
    axis = 1,
    how = 'any',
)
```

替换调NaN项

```python
df1 = df
df1.fillna(value='lsy')
```

判断是否有数据缺失

```python
df1 = df
df1.isnull()
np.any(df1.isnull()) == True  # 检查一堆数据里面是否有缺失
```
